```python
import pandas as pd 
import numpy as np 
kri_result=pd.read_csv('kri_result.csv')
kri_dictionary=pd.read_csv('kri_dictionary.csv')
Staff=pd.read_csv('staff.csv')
```

```python
def data_quality_kpis(response:dict):
    kri_dictionary['kri_id']=kri_dictionary['kri_id'].apply(lambda x: x.replace('KRI',''))
    kri_dictionary['kri_id']=kri_dictionary['kri_id'].astype(int)
    # ⚜️🐦‍🔥 new kri calculation from a kri_1 table:
    Kri_project_deadline=response

    #    {} 'kri':['project deadline','project deadline','project deadline'],
    #     'ggi':[100002,100003,100004],
    #     'snapshot_date':['2026-04-02','2026-04-01','2026-04-01']
    # }
    # affecting the kri to ⚜️🐦‍🔥the kri dictionary: 
    # initialization:
    kri_id=kri_dictionary['kri_id'].max()+1
    kri_group='project management'
    kri_label='project deadline'
    kri_dictionary_temp=pd.DataFrame()
    kri_dictionary_temp['kri_id']=[kri_id]
    kri_dictionary_temp['kri_group']=[kri_group]
    kri_dictionary_temp['kri_label']=[kri_label]
    kri_dictionary=pd.concat([kri_dictionary,kri_dictionary_temp],axis=0)

    # scalable data merging:
    Kri_project_deadline=pd.DataFrame(Kri_project_deadline)
    df=pd.merge(pd.merge(Staff,Kri_project_deadline, on='ggi', how='right'), kri_dictionary, left_on='kri', right_on='kri_label', how='left')
    kri_result=pd.concat([kri_result,df],axis=0)# ⚜️🐦‍🔥
# kri dataset:
```
